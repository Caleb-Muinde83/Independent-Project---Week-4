```python
#Importing the libraries to be used
#
import pandas as pd
import numpy as np
```

```python
# Loading the autolib dataset to be used in this analysis
# 
df = pd.read_csv("http://bit.ly/autolib_dataset")
df.head(10)
```

```python
#Accessing information about the data set
df.info()
```

### **Data Cleaning**

```python
# Irrelevant Data Observation
# Data Cleaning Action: Dropping geo point , displayed comment , and scheduled at attributes
# Explanation:No question to be answered requires this columns.
#
df.drop(["Geo point" , "Displayed comment" ,"Scheduled at"], axis = 1, inplace = True)
df.head(10)
```


**Combining Month, Year and Day columns with Pandas apply() and then use lambda function to combine the three values in a row using join() function.**

- we specify the columns of interest (Month, Year and Day columns) in our dataframe and then use lambda function to combine the three values in a row using join() function.

```python
# we specify the columns of interest (Month, Year and Day columns) in our dataframe 
# and then use lambda function to combine the three values in a row using join() function.
#
cols=["year","month","day"]
df['date'] = df[cols].apply(lambda x: '-'.join(x.values.astype(str)), axis="columns")
```

```python
#convert the date column into datetime type using Pandas to_datetime() function as shown in the post
#
df['date']=pd.to_datetime(df['date'])
```

```python
# Convert both columns to strings, add separator and use to_datetime with format parameter:
time = (pd.to_datetime(df['hour'].astype(str) + ':' + df['minute'].astype(str), format='%H:%M')
          .dt.time)
```

```python
# Creating the time column
#
df['time'] = time
df
```

```python
# Check if date column has been converted to date time format
df.info()
```

**The time column is in object format because it can not be converted to datetime data type.**

```python
#Data type conversion
#Converting the month,day,year,time columns to date time and combining them
df['Date'] =pd.to_datetime(df[['year', 'month', 'day', 'hour', 'minute']])
df
```

```python
#Dropping the minute,hour,day,month, and year columns now that we have a cobmined column
df.drop(["year" , "month" ,"day", "minute"], axis = 1, inplace = True)
df.head(10)
```

**Correcting spelling errors in Address and Public name colums from the dataset**

```python
# Syntax Errors -  Checking for syntax errors of leading or tailing white spaces in the dataset
#
df['Address'] = df['Address'].str.strip()
df['Address']
```

```python
# Removing ©- and ¨ then replacing "Ã" with "e"  the copyright sign from words in the address column
#
df['Address'] = df['Address'].str.replace("©","")
df['Address'] = df['Address'].str.replace("Ã","e")
df['Address'] = df['Address'].str.replace("¨","")
df['Address']
```

```python
# Syntax Errors -  Checking for syntax errors of leading or tailing white spaces in the dataset
#
df['Public name'] = df['Public name'].str.strip()
df['Public name']
```

```python
# Removing ©- and ¨ then replacing "Ã" with "e"  the copyright sign from words in the address column
#
df['Public name'] = df['Public name'].str.replace("©","")
df['Public name'] = df['Public name'].str.replace("Ã","e")
df['Public name'] = df['Public name'].str.replace("¨","")
df['Public name'] = df['Public name'].str.replace("","")
df['Public name'] = df['Public name'].str.replace("»","")
df['Public name']
```
